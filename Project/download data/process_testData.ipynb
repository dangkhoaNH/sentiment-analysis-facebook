```python
import pandas as pd
import regex as re
import emoji
import string
from underthesea import word_tokenize
from underthesea import sentiment
```

```python
test = pd.read_csv('test.csv')
test.rename(columns={'label':'emotion'}, inplace=True)
test
```

```python
# Basic preprocessing as lower string
def pre_processing(text):
  return text.str.lower()

test['comment'] = pre_processing(test['comment'])
test
```

```python
# Delete NaN and empty row
test = test[test.comment != '']
test = test.dropna(subset=['comment'])
test = test[test['comment'].apply(lambda comment: len(str(comment)) >= 5)]
test
```

```python
def remove_emoji(text):
  return emoji.get_emoji_regexp().sub("", text)

test['comment'] = test['comment'].map(lambda comment: remove_emoji(comment))
test
```

```python
test['comment'] = test['comment'].map(lambda comment: comment.replace(string.punctuation,''))
test['comment'] = test['comment'].map(lambda comment: comment.replace('_',' '))
test
```

```python
# Word segmentation
test['segment'] = test['comment'].map(lambda comment: word_tokenize(comment))
test
```

```python
# Remove vietnamese stopwords - https://github.com/stopwords/vietnamese-stopwords

# Read vietnamese stopwords file
stopwords_df = pd.read_table('vietnamese-stopwords.txt')
stopwords = stopwords_df.values.flatten()

# Remove vietnamese stopwords function
def remove_stopwords(segments):
  return [item for item in segments if item not in stopwords]

test['segment'] = test['segment'].map(lambda segments: remove_stopwords(segments))
test
```
