```python
import pandas as pd
import regex as re
import emoji
import string
from underthesea import word_tokenize
from underthesea import sentiment
```

```python
train = pd.read_csv('train.csv', encoding='utf-8')
train.rename(columns={'label':'emotion'}, inplace=True)
train
```

```python
# Basic preprocessing as lower string
def pre_processing(text):
  return text.str.lower()

train['comment'] = pre_processing(train['comment'])
train
```

```python
# Delete NaN and empty row
train = train[train.comment != '']
train = train.dropna(subset=['comment'])
train = train[train['comment'].apply(lambda comment: len(str(comment)) >= 5)]
train
```

```python
def remove_emoji(text):
  return emoji.get_emoji_regexp().sub("", text)

train['comment'] = train['comment'].map(lambda comment: remove_emoji(comment))
train
```

```python
train['comment'] = train['comment'].map(lambda comment: comment.replace(string.punctuation,''))
train['comment'] = train['comment'].map(lambda comment: comment.replace('_',' '))
train
```

```python
# Word segmentation
train['segment'] = train['comment'].map(lambda comment: word_tokenize(comment))
train
```

```python
# Remove vietnamese stopwords - https://github.com/stopwords/vietnamese-stopwords

# Read vietnamese stopwords file
stopwords_df = pd.read_table('vietnamese-stopwords.txt')
stopwords = stopwords_df.values.flatten()

# Remove vietnamese stopwords function
def remove_stopwords(segments):
  return [item for item in segments if item not in stopwords]

train['segment'] = train['segment'].map(lambda segments: remove_stopwords(segments))
train
```

```python
pd.Series(train['emotion'].values).value_counts()
```

```python
train.to_csv('comments-train.csv', index=False)
```
