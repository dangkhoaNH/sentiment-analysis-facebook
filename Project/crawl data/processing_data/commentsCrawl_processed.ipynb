```python
import pandas as pd
import regex as re
import emoji
import string
from underthesea import word_tokenize
from underthesea import sentiment
```

```python
# Read file crawlFB-80000-comments.csv

trainCrawl = pd.read_csv('crawlFB-80000-comments.csv')
trainCrawl
```

```python
# Delete index column

trainCrawl.drop(trainCrawl.columns[[0]], axis=1, inplace=True)
trainCrawl
```

```python
# Basic preprocessing as lower string
def pre_processing(text):
  return text.str.lower()

trainCrawl['comment'] = pre_processing(trainCrawl['comment'])
trainCrawl
```

```python
# Delete NaN and empty row
trainCrawl = trainCrawl[trainCrawl.comment != '']
trainCrawl = trainCrawl.dropna(subset=['comment'])
trainCrawl = trainCrawl[trainCrawl['comment'].apply(lambda comment: len(str(comment)) >= 5)]
trainCrawl
```

```python
def remove_emoji(text):
  return emoji.get_emoji_regexp().sub("", text)

trainCrawl['comment'] = trainCrawl['comment'].map(lambda comment: remove_emoji(comment))
trainCrawl
```

```python
trainCrawl['comment'] = trainCrawl['comment'].map(lambda comment: comment.replace(string.punctuation,''))
trainCrawl['comment'] = trainCrawl['comment'].map(lambda comment: comment.replace('_',' '))
trainCrawl
```

```python
# Word segmentation
trainCrawl['segment'] = trainCrawl['comment'].map(lambda comment: word_tokenize(comment))
trainCrawl
```

```python
# Remove vietnamese stopwords - https://github.com/stopwords/vietnamese-stopwords

# Read vietnamese stopwords file
stopwords_df = pd.read_table('vietnamese-stopwords.txt')
stopwords = stopwords_df.values.flatten()

# Remove vietnamese stopwords function
def remove_stopwords(segments):
  return [item for item in segments if item not in stopwords]

trainCrawl['segment'] = trainCrawl['segment'].map(lambda segments: remove_stopwords(segments))
trainCrawl
```

```python
def getEmotion(comment):
	emotion = sentiment(comment)
	if emotion is None:
		return 'neu'
	else:
		if emotion == 'positive':
			return 'pos'
		else: 
			return 'neg'
trainCrawl['emotion'] = trainCrawl['comment'].map(lambda comment: getEmotion(str(comment)))
trainCrawl
```

```python
pd.Series(trainCrawl['emotion'].values).value_counts()
```

```python
trainCrawl.to_csv('comments-'+str(trainCrawl.shape[0])+'-emotioned.csv', index=False)
```
